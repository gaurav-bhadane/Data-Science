### Feature Engineering

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
```

```python
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
df['Arrival_Hours']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minutes']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Arrival_Hours']=df['Arrival_Hours'].astype(int)
df['Arrival_Minutes']=df['Arrival_Minutes'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Departure_Hours']=df['Dep_Time'].str.split(':').str[0]
df['Departure_Minutes']=df['Dep_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Departure_Hours']=df['Departure_Hours'].astype(int)
df['Departure_Minutes']=df['Departure_Minutes'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4,np.nan:1})
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df.head()
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Hours']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df['Minutes']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.head()
```

```python
df['Hours'].isnull().sum()
```

```python
df['Minutes'].isnull().sum()
```

```python
df.info()
```

```python
df['Minutes'].fillna(0)
```

```python
df['Hours'] = df['Hours'].astype(str).str.extract(r'(\d+)').fillna(0).astype(int)
df['Minutes'] = df['Minutes'].astype(str).str.extract(r'(\d+)').fillna(0).astype(int)
```

```python
df['Total_Duration']=df['Hours']*60 + df['Minutes']
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('Duration',axis=1,inplace=True)
df.drop('Hours',axis=1,inplace=True)
df.drop('Minutes',axis=1,inplace=True)
```

### Converting Categorical Feature into Numerical Feature

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encode=OneHotEncoder()
```

```python
encoded_data=encode.fit_transform(df[['Airline','Source','Destination']]).toarray()
encoded_data
```

```python
df2=pd.DataFrame(encoded_data,columns=encode.get_feature_names_out())
```

```python
dataset = pd.concat([df, df2], axis=1)
```

```python
dataset
```

```python
dataset.drop('Airline',axis=1,inplace=True)
dataset.drop('Source',axis=1,inplace=True)
dataset.drop('Destination',axis=1,inplace=True)
```
