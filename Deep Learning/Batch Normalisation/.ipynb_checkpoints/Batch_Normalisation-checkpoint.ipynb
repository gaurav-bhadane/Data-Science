### Creating a Simple Classifier using Keras on MNIST data

```python
mnist=tf.keras.datasets.mnist
```

```python
mnist
```

```python
(X_train_full,y_train_full),(X_test,y_test)=mnist.load_data()
```

```python
X_train_full[0]
```

```python
X_train_full.shape
```

```python
X_test.shape
```

```python
X_train_full[0].shape
```

```python
X_train_full.dtype
```

```python
# Create Validation Dataset from the Full Training Data
# Scale the Data between 0 to 1 by dividing it by 255, as its an unsigned data between 0-255 range
X_valid,X_train=X_train_full[:5000]/255. , X_train_full[5000:]/255.
y_valid,y_train=y_train_full[:5000],y_train_full[5000:]

# Scale the Test set as well
X_test=X_test/255
```

```python
# Train: 55000
# Test: 10000
# Valid: 5000
```

```python
# Lets View Some Data

plt.imshow(X_train[1],cmap='binary')
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(X_train[1],annot=True,cmap='binary')
```

### Creating Layers of ANN

```python
LAYERS= [
    tf.keras.layers.Flatten(input_shape=[28,28],name="InputLayer"),
    tf.keras.layers.Dense(300,activation='relu',name="HiddenLayer1"),
    tf.keras.layers.Dense(100,activation='relu',name="HiddenLayer2"),
    tf.keras.layers.Dense(10,activation='softmax',name="OutputLayer")
]
```

```python
model_clf=tf.keras.models.Sequential(LAYERS)
```

```python
model_clf.layers
```

```python
model_clf.summary()
```

```python
hidden1=model_clf.layers[1]
hidden1.name
```

```python
hidden1.get_weights()[0].shape
```

```python
hidden1.get_weights()[1].shape
```

```python
weights,biases=hidden1.get_weights()
```

```python
weights.shape
```

```python
biases.shape
```

### LOSS Function

```python
LOSS_FUNCTION = 'sparse_categorical_crossentropy'
OPTIMIZER='SGD'
METRICS=["accuracy"]

model_clf.compile(loss=LOSS_FUNCTION,optimizer=OPTIMIZER,metrics=METRICS)
```

### Training

```python
EPOCHS=30
VALIDATION_SET=(X_valid,y_valid)

history=model_clf.fit(X_train,y_train,epochs=EPOCHS,validation_data=VALIDATION_SET,batch_size=32)
```

```python
history.params
```

```python
pd.DataFrame(history.history)
```

```python
pd.DataFrame(history.history).plot()
```

```python
model_clf.evaluate(X_test,y_test)
```

- In the abve Evaluate Method the 1st value is **Average Validation Loss** and 2nd value is **Average Validation Accuracy**
