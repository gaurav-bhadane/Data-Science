## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.random.rand(50)
```

```python
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
# Scatter Plot
plt.scatter(x,y)
```

```python
plt.scatter(x,y,c="red",alpha=0.8) #,marker="+"
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("X vs Y")
plt.grid()
plt.figure(figsize=(20,20))
```

```python
x=np.linspace(1,10,100)
y=np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y,"--r")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sin Wave")
```
