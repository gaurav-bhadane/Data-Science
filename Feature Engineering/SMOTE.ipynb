# SMOTE (Synthetic Minority Oversampling Techniques)
    SMOTE is a Technique used in Machine Learning to address imbalanced datasets where the minority class has significantly fewer instances than the Majority Class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,
                    weights=[0.9],random_state=12)
```

```python
x
```

```python
y
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
```

```python
df1
```

```python
df2
```

```python
df=pd.concat([df1,df2],axis=1)
```

```python
df
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['f1'],df['f2'],c=df['target'])
```

```python
pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
## Transform the Dataset
oversample=SMOTE()
x,y=oversample.fit_resample(df[['f1','f2']],df['target'])
```

```python
x.shape
```

```python
y.shape
```

```python
len(y[y==0])
```

```python
len(y[y==1])
```

```python
df3=pd.DataFrame(x,columns=['f1','f2'])
df4=pd.DataFrame(y,columns=['target'])
oversample_df=pd.concat([df3,df4],axis=1)
```

```python
oversample_df.head()
```

```python
oversample_df.shape
```

```python
plt.scatter(oversample_df['f1'],oversample_df['f2'],c=oversample_df['target'])
```

## Comparision Plot between Before Oversampling and After Oversampling

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].scatter(df['f1'], df['f2'], c=df['target'])
axs[0].set_title('Original Data')

axs[1].scatter(oversample_df['f1'], oversample_df['f2'], c=oversample_df['target'])
axs[1].set_title('Oversampled Data')

plt.show()
```
