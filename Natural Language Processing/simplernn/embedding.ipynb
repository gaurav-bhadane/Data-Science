```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good',
]
```

```python
sent
```

```python
# Define the vocabulary size
vocab_size=10000
```

```python
# One hot representation
one_hot_repr=[one_hot(word,vocab_size)for word in sent]
```

```python
one_hot_repr
```

```python
# word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# make every sentence of same length
max_len=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=max_len)
```

```python
embedded_docs
```

```python
## feature representation
dim=10
```

```python
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=dim, input_shape=(max_len,)))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
